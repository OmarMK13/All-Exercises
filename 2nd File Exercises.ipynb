```python
##Cards Game

from random import shuffle


class Card:
    categories = ["spades", "hearts", "diamonds","clubs"]

    values = ["2", "3",
              "4", "5", "6", "7",
              "8", "9", "10",
              "Jack", "Queen",
              "King", "Ace"]

    def __init__(self, v, s):
        self.value = v
        self.category = s

    def __lt__(self, c2):
        if self.value < c2.value:
            return True
        if self.value == c2.value:
            if self.category < c2.category:
                return True
            else:
                return False
        return False


    def __repr__(self):
        v = self.values[self.value] +\
            " of " + \
            self.categories[self.category]
        return v


class Deck:
    def __init__(self):
        self.cards = []
        for i in range(13):
            for j in range(4):
                self.cards.append(Card(i,j))
        shuffle(self.cards)

    def pop(self):
        if len(self.cards) == 0:
            return
        return self.cards.pop()


class Player:
    def __init__(self, name):
        self.wins = 0
        self.card = None
        self.name = name


class Game:
    def __init__(self):
        name1 = input("p1 name ")
        name2 = input("p2 name ")
        self.deck = Deck()
        self.p1 = Player(name1)
        self.p2 = Player(name2)

    def wins(self, winner):
        w = "{} wins this round"
        w = w.format(winner)
        print(w)

    def draw(self, p1n, p1c, p2n, p2c):
        d = "{} drew {} {} drew {}"
        d = d.format(p1n,
                     p1c,
                     p2n,
                     p2c)
        print(d)

    def play_game(self):
        cards = self.deck.cards
        print("Game Starts!!!")
        try:
            while len(cards) >= 2:
                m = "Type \'quit\' or \'play\'"
                response = input(m)
                if response == 'quit':
                    break
                elif response == 'play':    
                    p1c = self.deck.pop()
                    p2c = self.deck.pop()
                    p1n = self.p1.name
                    p2n = self.p2.name
                    self.draw(p1n,
                              p1c,
                              p2n,
                              p2c)
                    if p1c > p2c:
                        self.p1.wins += 1
                        self.wins(self.p1.name)
                    else:
                        self.p2.wins += 1
                        self.wins(self.p2.name)
                else:
                    raise Exception()
        except:
            print("Wrong Input")
            
            
        win = self.winner(self.p1,
                         self.p2)
        print("War is over.{} "
              .format(win))

    def winner(self, p1, p2):
        if p1.wins > p2.wins:
            return p1.name + " wins"
        if p1.wins < p2.wins:
            return p2.name + " wins"
        return "It was a tie!"

game = Game()
game.play_game()
```

```python
#Scraping Flipkart
from bs4 import BeautifulSoup
import requests


url = "https://www.flipkart.com/laptops/~buyback-guarantee-on-laptops-/pr?sid=6bo%2Cb5g&uniqBStoreParam1=val1&wid=11.productCard.PMU_V2"

response=requests.get(url)
soup=BeautifulSoup(response.content,'html.parser')

names = []
ratings = []
prices = []

for data in soup.findAll('div', class_='_3pLy-c row'):
        name=data.find('div',class_="_4rR01T")
        print(name.text)
        names.append(name.text)
        rating = data.find('div',class_="_3LWZlK")
        print(rating.text)
        ratings.append(rating.text)
        price=data.find('div',class_='_30jeq3 _1_WHN1')
        print(price.text)
        prices.append(price.text)
    
    

```
